## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#set bin parameters and names
bins = [0,9,14,19,24,29,34,39,150]
bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#ensure the SN is counted only once rather than each purchase
unique_players_df = purchase_data.drop_duplicates(subset=["SN"],keep="first",inplace=False)

#add bins to dataframe
unique_players_df["Age Group"] = pd.cut(unique_players_df["Age"], bins, labels=bin_names)

#Group by age and count total rows
grouped_unique_players_df = unique_players_df.groupby(["Age Group"])
count_grouped_unique_players_df = grouped_unique_players_df["Age"].count()
percent_grouped_unique_players = (count_grouped_unique_players_df/unique_players)*100

#create data frame and format etc
group_summary = pd.DataFrame({"Total Count":count_grouped_unique_players_df,
                             "Percentage of Players":percent_grouped_unique_players})

group_summary["Percentage of Players"] = group_summary["Percentage of Players"].map("{:.2f}%".format)
group_summary
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by both ID and name for each to be an index
grouped_item_df = purchase_data.groupby(["Item ID","Item Name"])

#assign variables and calculate (crucifer had 2 different prices so you have to convert the strings into floats as you need the decimals)
item_purchase_count = grouped_item_df["Purchase ID"].count()
grouped_item_price = grouped_item_df["Price"].unique()
grouped_item_price_final = grouped_item_price.astype(float)
grouped_item_total_price = grouped_item_df["Price"].sum()


Item_Summary = pd.DataFrame({"Purchase Count":item_purchase_count,
                             "Item Price":grouped_item_price_final,
                            "Total Purchase Value":grouped_item_total_price})
Item_Summary_Final = Item_Summary.sort_values("Purchase Count",ascending=False)
Item_Summary_Final["Total Purchase Value"] = Item_Summary_Final["Total Purchase Value"].map("${:.2f}".format)
Item_Summary_Final["Item Price"] = Item_Summary_Final["Item Price"].map("${:.2f}".format)
Item_Summary_Final.head()
```
